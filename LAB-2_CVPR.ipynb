```python
from cv2 import cvtColor,COLOR_BGR2RGB,COLOR_RGB2HSV,COLOR_RGB2LAB
```

```python
img_rgb=cvtColor(img,COLOR_BGR2RGB)
```

```python
print(img_rgb[0,0])
```

```python
plt.imshow(img_rgb)
```

```python
r,g,b=cv2.split(img_rgb)
```

```python
plt.figure(figsize=(15,10))
```

```python
plt.subplot(221)
plt.title("Original")
plt.imshow(img_rgb)
plt.subplot(222)
plt.title("Blue Channel")
plt.imshow(b)
plt.subplot(223)
plt.title("green channel")
plt.imshow(g)
plt.subplot(224)
plt.title("red channel")
plt.imshow(r)
```

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
plt.figure(figsize=(10,10)) 
```

```python
# Split HSV channels
h, s, v = cv2.split(img_hsv)

# Create subplots
plt.figure(figsize=(12, 6))
# Original Image
plt.subplot(221)
plt.title("Original")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')

# H Channel
plt.subplot(222)
plt.title("H Channel")
plt.imshow(h, cmap='gray')
plt.axis('off')

# S Channel
plt.subplot(223)
plt.title("S Channel")
plt.imshow(s, cmap='gray')
plt.axis('off')

# V Channel
plt.subplot(224)
plt.title("V Channel")
plt.imshow(v, cmap='gray')
plt.axis('off')

# Show the plot
plt.tight_layout()
plt.show()
```
